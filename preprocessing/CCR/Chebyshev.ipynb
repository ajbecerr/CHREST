```python
from io import StringIO
import numpy as np
```

```python
with open('Curran.txt', 'r') as soup:
    soup = soup.readlines()
    soup = soup[204:206]
    for i in range(len(soup)):
        soup[i] = soup[i].split(';')
        for j in range(len(soup[i])-1):
            soup[i][j] = soup[i][j].strip()
            soup[i][j] = soup[i][j].split('= ')[1]
            
Tlist = soup[0][0][1:-1].replace(',', '')
Plist = soup[0][1][1:-1].replace(',', '')
klist = soup[0][2][2:-2].replace(',', '')
klist = klist.replace('[', '')
klist = klist.replace(']', '\n')

Tlist = np.loadtxt(StringIO(Tlist))
Plist = np.loadtxt(StringIO(Plist))
klist = np.loadtxt(StringIO(klist))

print(Tlist)
print(Plist)
print(klist)
```

```python
np.save('Tlist.npy', Tlist)
np.save('Plist.npy', Plist)
np.save('klist.npy', klist)
```
